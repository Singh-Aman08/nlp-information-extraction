Sklearn Classification report

```python
def to_field(label):
    if label =="B-INT" or label =="I-INT":
        return "INT"
    elif label =="B-PART" or label == "I-PART":
        return "PART"
    elif label == "B-OUT" or label =="I-OUT":
        return "OUT"
    else:
        return "O"
```

```python
from sklearn.metrics import classification_report
```

```python
y_true = []
y_pred = []

for i in range(0,200):
  for t_doc, p_doc in zip(df['Merged BIO Labels'][i], df_few_shot['BIO Labels'][i]):
    y_true.append(to_field(t_doc))
    y_pred.append(to_field(p_doc))
```

```python
print(classification_report(y_true, y_pred))
```
